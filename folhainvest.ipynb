```python
import numpy as np
import pandas as pd
import os
import datetime
import scipy as sp
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#Lê todos arquivos
data = {}

for file in os.listdir('data'):
    table = pd.read_csv(os.path.join('data', file), decimal=",", sep='\t', header=1, encoding='latin-1')
    table.drop(['Nome', 'Última negociação', 'Hora da última negociação'], axis=1, inplace=True)
    
    #date = file
    date = datetime.datetime.strptime(file, '%Y%m%d')
    
    for i, _ in enumerate(table.loc[table.index]):
            try:
                data[table.loc[table.index[i]]['Ação']][date] = {
                    'Oscilação': table.loc[table.index[i]]['Oscilação'],
                    'Abertura': table.loc[table.index[i]]['Abertura'],
                    'Fechamento': table.loc[table.index[i]]['Fechamento'],
                    'Máxima': table.loc[table.index[i]]['Máxima'],
                    'Mínima': table.loc[table.index[i]]['Mínima'],
                    'Volume': table.loc[table.index[i]]['Volume de negociações']
                }
            except:
                data[table.loc[table.index[i]]['Ação']] = {}
```

```python
for d in data:
    print(d+'\n'+'*'*10)
    fig = plt.figure(figsize=(20, 20))
    r = data[d]
    r = sorted(r.items())

    for num, x in enumerate(r[0][1]):
        y = fig.add_subplot(5,4,num+1)
        ke = []
        de = []
        for l in r:
            ke.append(l[0])
            de.append(l[1][x])
        y.plot(ke, de)
        plt.title(x)
        #plt.xticks(rotation=90)
        y.axes.get_xaxis().set_visible(False)
        #y.axes.get_yaxis().set_visible(False)
    plt.show()
    #break
```
